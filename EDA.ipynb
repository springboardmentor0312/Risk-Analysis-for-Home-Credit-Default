```python
# Import necessary libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# Function to calculate Cramer's V
def cramers_v(chi2_stat, n, confusion_matrix):
    min_dim = min(confusion_matrix.shape) - 1
    return np.sqrt(chi2_stat / (n * min_dim))

# Step 1: Upload and load the dataset
print("Please upload your dataset file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]  # Dynamically get the uploaded file name
data = pd.read_csv(file_name)

# Step 2: Select categorical columns and encode them if necessary
categorical_cols = data.select_dtypes(include=['object']).columns

# Example: Calculate Chi-Square and visualize relationship
if 'home_ownership' in categorical_cols and 'Current_loan_status' in categorical_cols:
    # Create a contingency table
    contingency_table = pd.crosstab(data['home_ownership'], data['Current_loan_status'])

    # Perform Chi-Square test
    chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

    print(f"Chi-Square Test Statistics: {chi2_stat}")
    print(f"P-value: {p_val}")
    print(f"Degrees of Freedom: {dof}")

    if p_val < 0.05:
        print("There is a significant relationship between 'home_ownership' and 'Current_loan_status'.")
    else:
        print("There is no significant relationship between 'home_ownership' and 'Current_loan_status'.")

    # Cramr's V calculation
    n = contingency_table.sum().sum()  # Total number of observations
    cramer_v_value = cramers_v(chi2_stat, n, contingency_table)
    print(f"Cramr's V value: {cramer_v_value}")

    # Visualization: Stacked bar plot
    contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
    plt.title('Relationship Between Home Ownership and Current Loan Status', fontsize=16)
    plt.xlabel('Home Ownership', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.legend(title='Current Loan Status')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Either 'home_ownership' or 'Current_loan_status' is not present in the dataset.")
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
from scipy.stats import chi2_contingency

# Function to calculate Cramr's V statistic
def cramers_v(x, y):
    # Create a contingency table
    contingency_table = pd.crosstab(x, y)
    # Perform chi-squared test
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    # Calculate Cramr's V
    n = contingency_table.sum().sum()
    return np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

# Step 1: Upload and load the dataset
print("Please upload your dataset file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]  # Dynamically get the uploaded file name
data = pd.read_csv(file_name)

# Step 2: Select categorical features for correlation calculation
categorical_features = data.select_dtypes(include=['object', 'category'])  # Select only categorical columns

# Step 3: Create an empty correlation matrix
corr_matrix = pd.DataFrame(np.zeros((categorical_features.shape[1], categorical_features.shape[1])),
                           columns=categorical_features.columns,
                           index=categorical_features.columns)

# Step 4: Calculate Cramr's V for each pair of categorical features
for col1 in categorical_features.columns:
    for col2 in categorical_features.columns:
        corr_matrix.loc[col1, col2] = cramers_v(categorical_features[col1], categorical_features[col2])

# Step 5: Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# Title and labels
plt.title('Cramr\'s V Correlation Heatmap for Categorical Features', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/HCD(Preprocessed) (1).csv')  # Replace with the path to your dataset

# Ensure the 'home_ownership' and 'Current_loan_status' columns exist in your dataset
if 'home_ownership' in df.columns and 'Current_loan_status' in df.columns:
    # Calculate the percentage of defaulters by home_ownership
    home_ownership_status = df.groupby('home_ownership')['Current_loan_status'].value_counts(normalize=True).unstack().fillna(0)

    # Calculate percentage of defaulters (Current_loan_status == 'DEFAULT')
    home_ownership_status['DEFAULT_percentage'] = home_ownership_status['DEFAULT'] * 100

    # Plotting the bar graph
    home_ownership_status['DEFAULT_percentage'].plot(kind='bar', color='#FF6347', figsize=(12, 6))  # Tomato color for defaulters
    plt.title('Percentage of Defaulters by Home Ownership', fontsize=14)
    plt.xlabel('Home Ownership', fontsize=12)
    plt.ylabel('Percentage of Defaulters (%)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("The required columns ('home_ownership' and 'Current_loan_status') are not present in the dataset.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload and load the dataset
print("Please upload your dataset file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]  # Dynamically get the uploaded file name
data = pd.read_csv(file_name)

# Step 2: Check if 'loan_int_rate' and 'Current_loan_status' columns exist
if 'loan_int_rate' in data.columns and 'Current_loan_status' in data.columns:
    # Step 3: Bin 'loan_int_rate' into ranges
    rate_bins = pd.cut(data['loan_int_rate'], bins=[0, 5, 10, 15, 20, 25, 30],
                       labels=['0-5%', '5-10%', '10-15%', '15-20%', '20-25%', '25-30%'])

    # Step 4: Calculate the percentage of 'Current_loan_status' for each rate range
    rate_status_percentage = data.groupby(rate_bins)['Current_loan_status'].value_counts(normalize=True).unstack() * 100

    # Step 5: Plotting the percentage of 'Current_loan_status' by loan interest rate range
    rate_status_percentage.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#87CEEB', '#FFD700'])  # Light blue and gold
    plt.title('Percentage of Current Loan Status by Loan Interest Rate', fontsize=16)
    plt.xlabel('Loan Interest Rate (Ranges)', fontsize=12)
    plt.ylabel('Percentage (%)', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title='Current Loan Status', labels=['DEFAULT', 'NO DEFAULT'])
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("Columns 'loan_int_rate' or 'Current_loan_status' are not present in the dataset.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload and load the dataset
print("Please upload your dataset file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]  # Dynamically get the uploaded file name
data = pd.read_csv(file_name)

# Step 2: Check if the required columns exist
if 'term_years' in data.columns and 'Current_loan_status' in data.columns:
    # Step 3: Calculate the percentage of loan status for each term_years category
    term_status_percentage = data.groupby('term_years')['Current_loan_status'].value_counts(normalize=True).unstack() * 100

    # Step 4: Plot the percentage of Current Loan Status by Term Years
    term_status_percentage.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#87CEEB', '#FFD700'])  # Light blue and gold
    plt.title('Percentage of Current Loan Status by Term Years', fontsize=16)
    plt.xlabel('Term Years', fontsize=12)
    plt.ylabel('Percentage (%)', fontsize=12)
    plt.xticks(rotation=0)
    plt.legend(title='Current Loan Status', labels=['DEFAULT', 'NO DEFAULT'])
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("Columns 'term_years' or 'Current_loan_status' do not exist in the dataset.")
```
