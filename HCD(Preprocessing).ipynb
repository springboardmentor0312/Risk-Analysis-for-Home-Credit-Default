```python
df_cleaned = df.dropna(subset=['customer_id', 'loan_amnt', 'loan_int_rate', 'employment_duration', 'Current_loan_status'])
```

```python
df_cleaned.shape
```

```python
df_cleaned['customer_income'] = df_cleaned['customer_income'].replace('[£,]', '', regex=True).astype(float)
df_cleaned['loan_amnt'] = df_cleaned['loan_amnt'].replace('[£,]', '', regex=True).astype(float)
```

```python
df_cleaned['historical_default'] = df_cleaned['historical_default'].fillna('Unknown')
```

```python
df_cleaned.to_csv('/content/HCD(Preprocessed).csv', index=False)
```
